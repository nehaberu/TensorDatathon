```python
vict_r = [0,0,0,0,0]
print("VICTIMS: ")
for i in data.index:
    
    if data['Victim Race'][i] == 'Asian/Pacific Islander':
        vict_r[0] = vict_r[0]+1
    elif data['Victim Race'][i] == 'White':
        vict_r[1] = vict_r[1]+1
    elif data['Victim Race'][i] == 'Native American/Alaska Native':
        vict_r[2] = vict_r[2]+1
    elif data['Victim Race'][i] == 'Unknown':
        vict_r[3] = vict_r[3]+1
    elif data['Victim Race'][i] == 'Black':
        vict_r[4] = vict_r[4]+1    
    
print(vict_r)
```

```python
perp_r = [0,0,0,0,0]
print("PERPETRATORS: ")
for i in data.index:
    
    if data['Perpetrator Race'][i] == 'Asian/Pacific Islander':
        perp_r[0] = perp_r[0]+1
    elif data['Perpetrator Race'][i] == 'White':
        perp_r[1] = perp_r[1]+1
    elif data['Perpetrator Race'][i] == 'Native American/Alaska Native':
        perp_r[2] = perp_r[2]+1
    elif data['Perpetrator Race'][i] == 'Unknown':
        perp_r[3] = perp_r[3]+1
    elif data['Perpetrator Race'][i] == 'Black':
        perp_r[4] = perp_r[4]+1    
    
print(perp_r)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

GRAPHS SHOWING PROPORTIONS OF VICTIMS AND PERPETRATORS RACE-WISE

```python
label = ['Asian/Pacific Islander', 'White', 'Native American/Alaska Native', 'Unknown', 'Black']
vict_r = np.array(vict_r)
fig1, ax1 = plt.subplots()
explode = (0.05,0.05,0.05,0.2,0.05)
ax1.pie(vict_r, labels=label, autopct='%1.1f%%', explode=explode)
#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
label = ['Asian/Pacific Islander', 'White', 'Native American/Alaska Native', 'Unknown', 'Black']
perp_r = np.array(perp_r)
fig1, ax1 = plt.subplots()
explode = (0.05,0.05,0.05,0.05,0.05)
ax1.pie(perp_r, labels=label, autopct='%1.1f%%', explode=explode)
#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
